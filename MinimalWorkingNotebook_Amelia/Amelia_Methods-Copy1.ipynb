```python
def Disperse(self):
    sig = self.signal[1] 
    DM = self.DM
    f0 = self.f0
    dt = self.TimeBinSize
    fourier = np.fft.rfft(sig)
    #length = len(fourier)
    #values = np.linspace(-100,100,length) #Would it be useful to change -100,100 to something else?
    #freqs = np.fft.rfftfreq(2*len(values)-1,d=dt/1e6)*1e3
    freqs = np.fft.rfftfreq(2*len(fourier)-1,d=dt/1e6)*1e3
    FinalFreqs = freqs-f0+1e-10 # Added the 1e-10 to avoid division by 0 errors in exponent
    #print FinalFreqs
    H = np.exp(1j*2*np.pi*4.148808e9/((FinalFreqs+f0)*f0**2)*DM*FinalFreqs**2)
    product = fourier*H
    Dispersed = np.fft.irfft(product)
    plot(sig+30)
    plot(Dispersed)
    show()
    return Dispersed
```

```python
def plot_dispersed(pulsar_object, ism_object, signal_object):
    lim = pulsar_object.T/pulsar_object.Signal_in.TimeBinSize
    print lim
    if signal_object.SignalType == 'voltage':
        title('Voltage vs. Time')
        ylabel('Voltage')
        xlabel('Time (ms)')
        xlim(0,lim)
    elif signal_object.SignalType == 'intensity':
        title('Intensity vs. Time')
        ylabel('Intensity')
        xlabel('Time (ms)')
        xlim(0,lim)
    plot(Disperse(ism_object))
    show()
```
