# Let's Convolve Things

```python
top_hat = signal.boxcar(499)/499
a=np.max(top_hat)
print(a)
```

Add some noise

```python
noisy_gauss=(P1.profile[3,:])*(np.random.normal(0,1,499)**2)
noisy_tophat=top_hat*(np.random.normal(0,1,499)**2)
```

```python
plt.plot(P1.profile[3,:], label="Profile")
plt.plot(noisy_gauss, label="Noisy Profile")
plt.xlim(70,130)
plt.legend()
```

```python
convoled_noisy_gauss=8*np.convolve(noisy_tophat,P1.profile[3,:]'same')
```

```python
plt.plot(P1.profile[3,:], label="Profile")
plt.plot(noisy_gauss, label="Noisy Profile")
plt.plot(convoled_noisy_gauss, label="Convolved Noisy Profile")
#plt.xlim(70,130)
plt.legend()
```
